```python
# you will need the following libraries
# pip install pyexcel==0.5.15
# pip install pyexcel-text
# pip install pyecharts==1.3.0
import pyexcel as p

sheet = p.get_sheet(url='https://www.worldometers.info/world-population/population-by-country/')
```

```python
sheet
```

```python
sheet.column.select([1,2])
```

```python
data = sheet.to_array()
data
```

```python
high = max([x for _, x in data[1:]])
low = min([x for _, x in data[1:]])
```

```python
from pyecharts.charts import Map
import pyecharts.options as opts

m = (Map().add(
    maptype='world',
    series_name="World Population",
    data_pair=data[1:],
    is_map_symbol_show=False,
    label_opts=opts.LabelOpts(is_show=False),
))
m.render_notebook()
```

```python
m.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(
        min_=low,
        max_=high,
        range_text=["max", "min"],
        is_calculable=True,
        range_color=["lightskyblue", "yellow", "orangered"],
    )
)
m.render_notebook()
```

```python
m = (Map().add(
    maptype='world',
    series_name="World Population",
    data_pair=data[1:],
    is_map_symbol_show=False,
    label_opts=opts.LabelOpts(is_show=False),
    name_map={
        'Dem. Rep. Congo': 'DR Congo',
        'Central African Rep.': 'Central African Republic',
        'S. Sudan': 'South Sudan',
        'Korea': 'South Korea'
    }
).set_global_opts(visualmap_opts=opts.VisualMapOpts(
        min_=low,
        max_=high,
        range_text=["max", "min"],
        is_calculable=True,
        range_color=["lightskyblue", "yellow", "orangered"],
    )
))
m.render_notebook()
```
